```python
import pandas as pd
```

```python
df=pd.read_csv('newmonthsdb.csv', index_col=0)
```

```python
df
```

```python
df['created_at_month'].value_counts()
```

### Dinero total solicitado cada mes.

```python
df_eu_mes=df.groupby(['created_at_month'])['amount'].sum()
```

```python
df_eu_mes.to_frame().reset_index()
df_eu_mes
```

### Dinero medio solicitado por cada usuario al mes

```python
dfmean1=df.groupby(['created_at_month'])['user_id'].count()
```

```python
dfmean1
```

```python
df_tot=df.groupby(['created_at_month'])['amount'].sum()
```

```python
result = pd.concat([dfmean1, df_tot], axis=1)
result
```

```python
result['mean']=result['amount']/result['user_id']
```

```python
result.drop(['user_id','mean'], axis=1,inplace=True)
result
```

### Número de solicitudes al mes que son aceptadas pero NO son devueltas

```python
dfacc=df[df['status']=='approved']
```

```python
dfacc=dfacc.groupby(['created_at_month']).count()
```

```python
dfacc
```

```python
dfacc['nodevueltas']=dfacc['transfer_type']-dfacc['money_back_date']
df_nodevueltas=dfacc['nodevueltas']
df_nodevueltas.columns =['Month', 'NoDevueltas_total']
```

```python
df_nodevueltas
```

### Número de solicitudes al mes que son aceptadas y Sí son devueltas

```python
df_devueltas=dfacc.copy().reset_index()
```

```python
df_devueltas
```

```python
df_devueltas=df_devueltas[['created_at_month','money_back_date']]
```

```python
df_devueltas.columns =['Month', 'Devueltas_total']
```
